```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/tanzania.csv")
df.head()
```

```python
df["Country"] = "Tanzania"
df.head()
```

```python
df["DATE"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df.head()
```

```python
df["Month"] = df["DATE"].dt.month
df.head()
```

```python
df = df.replace(-999, np.nan)
```

```python
df.describe()
```

Summary Statistics Interpretation

- Temperature is high throughout the dataset, with an average around 27°C, indicating a consistently warm tropical climate.
- Maximum temperatures remain relatively stable, showing low extreme cold variation.
- Precipitation shows variability, with occasional high rainfall events compared to generally moderate dry periods.
- Humidity levels are high (~80%), indicating a very moist climate environment.
- Wind speeds are moderate and stable, with limited extreme variation.
- Atmospheric pressure remains steady, showing no strong variability patterns.

```python
missing_count = df.isna().sum()
missing_percent = (missing_count / len(df)) * 100

missing_report = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing %": missing_percent
})

missing_report
```

```python
from scipy.stats import zscore

cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

z = df[cols].apply(zscore)

outlier_rows = (np.abs(z) > 3).any(axis=1)
outlier_count = outlier_rows.sum()
outlier_count
df[outlier_rows]
```

Outlier Detection (Z-Score Method)

- Z-score analysis was applied to key climate variables including temperature, precipitation, humidity, and wind speed.
- Observations exceeding |3| standard deviations were flagged as outliers.
- A small number of extreme values were detected, mainly driven by precipitation spikes.
- These outliers are likely real weather extremes rather than data errors.
- Therefore, they were retained to preserve climate variability.

```python
row_missing_pct = df.isna().mean(axis=1) * 100
row_missing_pct.describe()
```

```python
df = df[row_missing_pct <= 30]
```

```python
df = df.sort_values("DATE")
df = df.ffill()
```

```python
df = df.drop_duplicates()
```

Missing Value Handling Decision (Kenya)

Rows with more than 30% missing values were removed to ensure data quality and avoid unreliable records.

Remaining missing values were handled using forward-fill imputation, which is appropriate for time-series weather data because it preserves temporal continuity and avoids breaking seasonal patterns.

This approach maintains the natural structure of the climate dataset while ensuring smooth and consistent trends without introducing artificial bias.

```python
df.to_csv("../data/tanzania_clean.csv", index=False)
```

```python
monthly_temp = df.groupby("Month")["T2M"].mean()
monthly_temp
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(monthly_temp.index, monthly_temp.values)
plt.title("Monthly Average Temperature (T2M)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```

Temperature Trend Analysis

- The highest average temperature occurs around March and April.
- The lowest average temperature occurs around the month of July.
- The pattern shows a clear seasonal trend, with temperatures increasing and decreasing throughout the year.
-This indicates clear climatic seasonality in Kenya, with warmer and cooler periods

```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()
monthly_rain
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(monthly_rain.index, monthly_rain.values)
plt.title("Monthly Total Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

Rainfall Analysis

The highest rainfall occurs in March,April and May indicating the peak rainy season.
Rainfall is also relatively high in  November,December and January, showing an extended wet period.
The lowest rainfall occurs in June, July,August and September, indicating dry season conditions.
The rainfall pattern is strongly seasonal, with a clear distinction between wet and dry months.
