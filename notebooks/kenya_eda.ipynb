# Kenya Climate EDA

NASA Climate Dataset for Kenya (2015–2026)

### Objective
Analyze, clean, and explore Kenya’s climate data to identify trends, variability, and climate risks relevant to COP32.

```python
import pandas as pd
import numpy as np
```

### 1. Data Loading and Date Parsing

```python
df = pd.read_csv("../data/kenya.csv")
df.head()
```

```python
df["Country"] = "Kenya"
df.head()
```

### 2. Summary Statistics and Missing Value Report

```python
df["DATE"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df.head()
```

```python
df["Month"] = df["DATE"].dt.month
df.head()
```

```python
df = df.replace(-999, np.nan)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

Summary Statistics Interpretation

- Temperature values show a moderate climate, with mean T2M around 20.4°C and maximum values reaching ~27.8°C.
- Temperature variability is moderate, with clear differences between minimum and maximum temperatures.

- Precipitation is generally low (mean ~1.47 mm), indicating frequent dry conditions with occasional rainfall events.

- Humidity levels are relatively high (mean ~65%), suggesting a generally moist environment.

- Wind speeds are moderate (~3 m/s) with occasional stronger events up to ~4.3 m/s.

- Atmospheric pressure remains stable, showing minimal variation across the dataset.

### 3. Outlier Detection and Basic Cleaning

```python
missing_count = df.isna().sum()
missing_percent = (missing_count / len(df)) * 100

missing_report = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing %": missing_percent
})

missing_report
```

```python
from scipy.stats import zscore

cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

z = df[cols].apply(zscore)

outlier_rows = (np.abs(z) > 3).any(axis=1)
outlier_count = outlier_rows.sum()
outlier_count
df[outlier_rows]
```

Outlier Detection (Z-Score Method)

- Z-score analysis was applied to key climate variables (temperature, precipitation, humidity, and wind speed).
- Rows with values greater than |3| standard deviations were flagged as outliers.

- A total number of extreme observations were identified, mainly driven by precipitation and temperature variability.
- These outliers likely represent real extreme weather events rather than data errors.

- Therefore, they are retained to preserve climate variability and extreme event information.

```python
row_missing_pct = df.isna().mean(axis=1) * 100
row_missing_pct.describe()
```

```python
df = df[row_missing_pct <= 30]
```

```python
df = df.sort_values("DATE")
df = df.ffill()
```

```python
df = df.drop_duplicates()
```

Missing Value Handling Decision 
Rows with more than 30% missing values were removed to ensure data quality and avoid unreliable records.

Remaining missing values were handled using forward-fill imputation, which is appropriate for time-series weather data because it preserves temporal continuity and avoids breaking seasonal patterns.

This approach maintains the natural structure of the climate dataset while ensuring smooth and consistent trends without introducing artificial bias.

```python
df.to_csv("../data/kenya_clean.csv", index=False)
```

### 4. Time Series Analysis

```python
monthly_temp = df.groupby("Month")["T2M"].mean()
monthly_temp
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(monthly_temp.index, monthly_temp.values)
plt.title("Monthly Average Temperature (T2M)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```

Temperature Trend Analysis

- The highest average temperature occurs around March and April.
- The lowest average temperature occurs around the month of July.
- The pattern shows a clear seasonal trend, with temperatures increasing and decreasing throughout the year.
-This indicates clear climatic seasonality in Kenya, with warmer and cooler periods

```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()
monthly_rain
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(monthly_rain.index, monthly_rain.values)
plt.title("Monthly Total Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

Rainfall Analysis

The highest rainfall occurs in March,April and May indicating the peak rainy season.
Rainfall is also relatively high in  November and December, showing an extended wet period.
The lowest rainfall occurs in June, July,August and November, indicating dry season conditions.
The rainfall pattern is strongly seasonal, with a clear distinction between wet and dry months.

### 5. Correlation and Relationship Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

 Correlation Analysis
- Strong positive correlations are observed between temperature variables (T2M, T2M_MAX, T2M_MIN), indicating they move together consistently.
- Wind variables (WS2M, WS2M_MAX) show moderate positive relationships, suggesting relatively stable wind behavior patterns.
- Temperature and humidity show a weak negative relationship, which is expected in typical climate systems.
- Overall, the correlations reflect realistic atmospheric interactions and consistent climate behavior in the dataset.
