```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/nigeria.csv")
df.head()
```

```python
df["Country"] = "Nigeria"
df.head()
```

```python
df["DATE"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df.head()
```

```python
df["Month"] = df["DATE"].dt.month
df.head()
```

```python
df = df.replace(-999, np.nan)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

Summary Statistics Interpretation 
- Temperatures are consistently high, with a mean of ~26.7°C, indicating a warm tropical climate.
- Maximum temperatures reach ~28.9°C, while minimum values remain relatively high (~24.9°C), showing low temperature variability.

- Precipitation is moderate on average (~4.21 mm), but includes extreme spikes up to 166 mm, indicating occasional heavy rainfall events.

- Humidity levels are very high (~85%), suggesting a persistently moist and humid environment.

- Wind speeds are low to moderate (~2.2 m/s), with limited extreme variation.

- Atmospheric pressure is stable around ~100.8, showing minimal variability across the dataset.

```python
missing_count = df.isna().sum()
missing_percent = (missing_count / len(df)) * 100

missing_report = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing %": missing_percent
})

missing_report
```

```python
from scipy.stats import zscore

cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

z = df[cols].apply(zscore)

outlier_rows = (np.abs(z) > 3).any(axis=1)
outlier_count = outlier_rows.sum()
outlier_count
df[outlier_rows]
```

Outlier Detection (Z-Score Method)

- Z-score analysis was applied to key climate variables including temperature, precipitation, humidity, and wind speed.
- Observations exceeding |3| standard deviations were flagged as potential outliers.

- Several extreme values were detected, mainly driven by unusually high rainfall events and occasional temperature variations.
- The highest outliers correspond to intense rainfall spikes (e.g., above 30 mm), which reflect real weather events.
- These outliers are considered valid climatic extremes rather than data errors.
- Therefore, they were retained to preserve the natural variability and extreme weather behavior of Nigeria’s climate.

```python
row_missing_pct = df.isna().mean(axis=1) * 100
row_missing_pct.describe()
```

```python
df = df[row_missing_pct <= 30]
```

```python
df = df.sort_values("DATE")
df = df.ffill()
```

```python
df = df.drop_duplicates()
```

Missing Value Handling Decision
Rows with more than 30% missing values were removed to ensure data quality and avoid unreliable records.

Remaining missing values were handled using forward-fill imputation, which is appropriate for time-series weather data because it preserves temporal continuity and maintains the structure of seasonal climate patterns.

This approach ensures that Nigeria’s highly variable climate characteristics, including rainfall spikes and humidity fluctuations, are preserved while maintaining smooth and consistent trends without introducing artificial bias.

```python
df.to_csv("../data/nigeria_clean.csv", index=False)
```

```python
monthly_temp = df.groupby("Month")["T2M"].mean()
monthly_temp
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(monthly_temp.index, monthly_temp.values)
plt.title("Monthly Average Temperature (T2M)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```

Temperature Trend Analysis

- The highest average temperature occurs around March and April.
- The lowest average temperature occurs around the month of August.
- The pattern shows a clear seasonal trend, with temperatures increasing and decreasing throughout the year.
-This indicates clear climatic seasonality in Kenya, with warmer and cooler periods

```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()
monthly_rain
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(monthly_rain.index, monthly_rain.values)
plt.title("Monthly Total Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

Rainfall Analysis

The highest rainfall occurs in June indicating the peak rainy season.
Rainfall is also relatively high in  September and october, showing an extended wet period.
The lowest rainfall occurs in December, January and February, indicating dry season conditions.
The rainfall pattern is strongly seasonal, with a clear distinction between wet and dry months having high season of rainfall.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Correlation Analysis

- Strong positive correlations are observed among temperature variables (T2M, T2M_MAX, T2M_MIN), showing consistent temperature behavior across the dataset.
- Humidity shows a weak to moderate relationship with temperature, reflecting typical tropical interactions where moisture levels vary with heat.
- Wind speed shows weak correlations with other variables, indicating largely independent atmospheric behavior.
- Rainfall has weak and inconsistent correlations due to its irregular and extreme event-driven nature.
- Overall, the correlation structure reflects a stable tropical climate system with rainfall-driven variability.
